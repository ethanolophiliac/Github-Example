```python
import requests
import lxml.html as lh
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

page = requests.get(url)

doc = lh.fromstring(page.content)

elements = doc.xpath('//tr')

[len(row) for row in elements[:10]]
```

```python
empty = []
i = 0

for row in elements[0]:
    i+=1
    name = row.text_content()
    empty.append((name,[]))
```

```python
for i in range (1, len(elements)):
    number = elements[i]
    
    if len(number) != 3:
        break
    
    i = 0
    
    for t in number.iterchildren():
        data = t.text_content()
        if i > 0:
            try:
                data=int(data)
            except:
                pass
        empty[i][1].append(data)
        i+=1
```

```python
Dict = {title:column for (title,column) in empty}
df = pd.DataFrame(Dict)
df.columns = ['Postcode','Borough', 'Neighbourhood']
df.head()
```

```python
df = df.replace('\n', '', regex = True)
df.head()
```

```python
df.drop(df.index[df['Borough'] == 'Not assigned'], inplace = True)
df.reset_index(drop=True, inplace=True)
df
```

```python
url = 'http://cocl.us/Geospatial_data'

df1 = pd.read_csv(url)

df1.head()
```

```python
df1.columns = ['Postcode', 'Latitude', 'Longitude']
df1.head()
```

```python
df_new = pd.merge(df, df1, on='Postcode')
df_new.head()
```

```python
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
        
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID,
            CLIENT_SECRET,
            VERSION,
            lat,
            lng,
            radius,
            LIMIT)
        
        results= requests.get(url).json()["response"]['groups'][0]['items']
    
        venues_list.append([(
        name,
        lat,
        lng,
        v['venue']['name'],
        v['venue']['location']['lat'],
        v['venue']['location']['lng'],
        v['venue']['categories'][0]['name'])  for v in results])
    
    nearby_venues=pd.DataFrame([item for venues_list in venues_list for item in venues_list])
    
    nearby_venues.columns = ['Neighborhood',
                            'Neighborhood Latitude',
                            'Neighborhood Longitude',
                            'Venue',
                            'Venue Latitude',
                            'Venue Longitude',
                            'Venue Category']

    return(nearby_venues)
                                                        
```

**Analyze and Group Each Neighborhood

```python
toronto_encoded = pd.get_dummies(toronto_venues[['Venue Category']], prefix='', prefix_sep='')

toronto_encoded['Neighborhood'] = toronto_venues['Neighborhood']

fixed_columns = [toronto_encoded.columns[-1]] + list(toronto_encoded.columns[:-1])
toronto_encoded = toronto_encoded[fixed_columns]

toronto_encoded.head()
```

```python
toronto_grouped = toronto_encoded.groupby('Neighborhood').mean().reset_index()
toronto_grouped.head()
```

**Top 10 Venues

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[2:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

```python
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

columns = ['Neighborhood']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

neighborhoods_venues_sorted = pd.DataFrame(columns=columns)
neighborhoods_venues_sorted['Neighborhood'] = toronto_grouped['Neighborhood']

for ind in np.arange(toronto_grouped.shape[0]):
    neighborhoods_venues_sorted.iloc[ind, 1:] = return_most_common_venues(toronto_grouped.iloc[ind, :], num_top_venues)

neighborhoods_venues_sorted.head()
```

**Clustering Time 

```python
toronto_grouped_cluster = toronto_grouped.drop('Neighborhood',1)
kclusters = 5

kmeans = KMeans(n_clusters = kclusters, random_state=0).fit(toronto_grouped_cluster)

kmeans.labels_[0:10]
```

```python
neighborhoods_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)

toronto_merged = df_new

toronto_merged = toronto_merged.join(neighborhoods_venues_sorted.set_index('Neighborhood'),
                                     on='Neighbourhood')
toronto_merged.head()
```
